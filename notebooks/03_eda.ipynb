# 03 Exploratory Data Analysis (EDA)

This notebook performs EDA on the processed Iris data and generates a pairplot to visualize feature relationships.

```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join('..', 'src')))
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ds_logger import start_logging, end_logging

notebook_description = "Performs EDA on the processed Iris data and saves a pairplot."
start_logging(notebook_name='03_eda.ipynb', notebook_description=notebook_description)
```

```python
# Find the latest processed data file
processed_data_dir = '../data/processed'
latest_processed_file = max([os.path.join(processed_data_dir, f) for f in os.listdir(processed_data_dir) if f.startswith('iris_processed') and f.endswith('.csv')], key=os.path.getctime)
df = pd.read_csv(latest_processed_file)

# Generate pairplot
sns.pairplot(df, hue='target')

# Save plot
plot_path = '../reports/iris_pairplot.png'
plt.savefig(plot_path)
```

```python
end_logging(results={'plot_generated': 'iris_pairplot.png'})
```
